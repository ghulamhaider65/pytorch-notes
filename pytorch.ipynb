```python
import time
 
# Create tensors on CPU
a_cpu = torch.rand((10000, 10000))
b_cpu = torch.rand((10000, 10000))
 
# Measure time for CPU multiplication
start_cpu = time.time()
result_cpu = torch.matmul(a_cpu, b_cpu)
end_cpu = time.time()
print(f"CPU multiplication time: {end_cpu - start_cpu} seconds")
 
# Create tensors on GPU
a_gpu = torch.rand((10000, 10000), device='cuda')
b_gpu = torch.rand((10000, 10000), device='cuda')
 
# Measure time for GPU multiplication
start_gpu = time.time()
result_gpu = torch.matmul(a_gpu, b_gpu)
end_gpu = time.time()
print(f"GPU multiplication time: {end_gpu - start_gpu} seconds")
```
